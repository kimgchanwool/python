```python
import time

s_t = time.time()
count = 0
def recursive_function():
    global count
    count +=1
    print('재귀 함수를 호출합니다.', count)
    e_t = time.time()
    print('현 시각은', e_t-s_t)
    recursive_function()

recursive_function()
```

```python
def recursive_function(i):
    #100번 호출을 했을 때 종료되도록 종료 조건 명시
    if i == 100:
        return
    print(i, '번째 재귀함수에서', i+1, '번째 재귀함수를 호출합니다.')
    recursive_function(i+1)
    print(i, '번째 재귀함수를 종료합니다.')
recursive_function(1)
```

### 호출 종료시점은 함수에서 재귀부른 함수가 끝나야하므로 1번째가 가장 나중에 종료된다.

```python
def recursive_factorial(i):
    if i <= 0:
        return 1
    return recursive_factorial(i-1) * i
recursive_factorial(6)
```

```python
def factorial(i):
    num = 1
    while i > 0:
        num *= i
        i-=1
    return(num)
factorial(6)
```

```python
#이진 탐색 라이브러리
from bisect import bisect_left, bisect_right, bisect

a = [1, 2, 4, 4, 8]
x = 4

print(bisect_left(a, x)) #x가 들어갈 수 있는 자리중 가장 좌측인덱스를 출력
print(bisect_right(a, x)) #x가 들어갈 수 있는 자리 중 가장 우측 인덱스를 출력
print(bisect(a, x)) #default는 우측.
```

```python
# 범위에 속하는 데이터 개수 구하기
def count_by_range(a, left_value, right_value): 
    left_index =  bisect_left(a, left_value) #원하는 데이터의 제일 좌측
    right_index = bisect_right(a, right_value) #원하는 데이터의 제일 우측
    return right_index - left_index # 우측 - 좌측은 원하는 범위에 있는 데이터 개수
```

```python
a = [1, 2 ,3, 3, 3, 3, 4, 4, 8, 9]
print(count_by_range(a, 4, 4)) #4부터 4 사이에 있는 데이터의 개수
print(count_by_range(a, -1, 3)) #-1부터 3 사이에 있는 데이터의 개수
```

```python
def bisect_count(a, value):
    end = len(a) - 1
    count1 = count2 = 0
    for i in range(len(a)):
        if a[i] == value and count1 == 0:
            count1 = 1
            st = i
        if a[end-i] == value and count2 == 0:
            count2 = 1
            en = end - i
        if count2 == 1 and count1 == 1:
            return en - st + 1
    return -1
print(a)
print(bisect_count(a, 3))
#함수 없이
```
