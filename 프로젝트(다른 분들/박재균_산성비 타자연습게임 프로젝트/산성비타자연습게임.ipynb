```python
import pygame
import random

#코드 선택
def code_selection(): #단어리스트1
    code_list = open('words.txt').read().splitlines()
    
    elements_cnt = len(code_list)
    i = random.randint(0, elements_cnt - 1)
    return code_list[i]

def word_selection(): #단어리스트2
    word_list = open('words1.txt').read().splitlines()
    
    elements_cnt = len(word_list)
    i = random.randint(0, elements_cnt - 1)
    return word_list[i]

def phrase_selection(): #단어리스트3
    word_list = open('words2.txt').read().splitlines()
    
    elements_cnt = len(word_list)
    i = random.randint(0, elements_cnt - 1)
    return word_list[i]

pygame.init() # 초기화

screen_width = 640
screen_height = 960

screen = pygame.display.set_mode((screen_width, screen_height)) # 디스플레이 설정
font_rain = pygame.font.SysFont('Arial', 30)
font_big = pygame.font.SysFont('Arial Black', 80)
pygame.display.set_caption('산성비 타자연습 게임') # 타이틀

## 사용자 설정

word = code_selection() # 단어1
word1 = word_selection() #단어2
word2 = phrase_selection() #단어3
userinput = '' #입력창
score = 0 #점수
stage = 1
font = pygame.font.SysFont('Britannic Bold',32) #폰트
start_ticks = pygame.time.get_ticks()
green = (30, 150, 30)
red = (150, 50, 50)
yellow = (150, 150, 30)

#산성비 만들기
rain = word
rain_x_pos = random.randint(0, 450)
rain_y_pos = 0
rain_speed = 0.03

rain1 = word1
rain_x_pos1 = random.randint(0, 450)
rain_y_pos1 = 0
rain_speed1 = 0.05

rain2 = word2
rain_x_pos2 = random.randint(0, 450)
rain_y_pos2 = 0
rain_speed2 = 0.07

    
running = True

while running:
    seconds= (pygame.time.get_ticks() - start_ticks)/1000
# - pygame.time.get_ticks()) #start_ticks
    for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif seconds >= 33:
                running = False
            elif stage == 30:
                running = False
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_RETURN: #엔터키눌렀을때
                    
                    if userinput == word :
                        stage += 1
                        word = code_selection()
                        userinput =''
                        score += 30
                        rain_y_pos = 0
                        rain_x_pos = random.randint(0, 450)

                        
                    if userinput == word1 :
                        rain_y_pos1 = 0
                        stage += 1
                        word1 = word_selection()
                        userinput =''
                        score += 40
                        rain_y_pos1 = 0
                        rain_x_pos1 = random.randint(0, 450)
                        
                        
                    if userinput == word2 :
                        rain_ypos2 = 0
                        stage += 1
                        word2 = phrase_selection()
                        userinput =''
                        score += 50
                        rain_y_pos2 = 0
                        rain_x_pos2 = random.randint(0, 450)
                        

                    else:
                        userinput=''
                        #score -= 50
                        
                elif event.key == pygame.K_BACKSPACE: #백스페이스
                    userinput = userinput[:-1]

                else :
                    userinput += chr(event.key) #이벤트키 문자화
                    
                
    
    #산성비 위치
    rain_y_pos += rain_speed
    
    if rain_y_pos > 960:    #y축 다시시작
        rain_y_pos = 0
        rain_x_pos = random.randint(0, 450)
        score -=30 #바닥으로 떨어지면 -30점
        
    rain_y_pos1 += rain_speed1
    
    if rain_y_pos1 > 960:    
        rain_y_pos1 = 0
        rain_x_pos1 = random.randint(0, 450)
        score -=30 
    
    rain_y_pos2 += rain_speed2
    
    if rain_y_pos2 > 960:   
        rain_y_pos2 = 0
        rain_x_pos2 = random.randint(0, 450)
        score -=30 
        
                    
    screen.fill((200, 200, 200))
    
    sf_word = font_rain.render(word, True, (0, 255, 255))
    sf_word1 = font_rain.render(word1, True, (255, 0, 255))
    sf_word2 = font_rain.render(word2, True, (255, 255, 0))
    

 
    screen.blit(sf_word, (rain_x_pos, rain_y_pos))
    screen.blit(sf_word1, (rain_x_pos1, rain_y_pos1))
    screen.blit(sf_word2, (rain_x_pos2, rain_y_pos2))

    stageCaption = font.render('Word Count : ' +str(stage), True, green)
    timerCaption = font.render('Time : ' +str(seconds),True, red)
    userInputCaption = font.render('Input : ' + str(userinput),True, red)
    scoreCaption = font.render('Score : '+str(score),True, yellow)
    screen.blit(stageCaption,(5,10))
    screen.blit(timerCaption,(250,10))
    screen.blit(userInputCaption,(100,900))
    screen.blit(scoreCaption,(500,10))
    
    
    if seconds >= 30: 
        gmovr = font_big.render('Game Over', True, red)
        scr = font_big.render('SCORE: {}'.format(score), True, green)
        screen.blit(gmovr, gmovr.get_rect(center=(screen_width/2, screen_height/2-75)))
        screen.blit(scr, scr.get_rect(center=(screen_width/2, screen_height/2+75)))
    

    pygame.display.update()

pygame.quit()
```
