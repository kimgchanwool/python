```python
from selenium import webdriver
import time
import requests
from bs4 import BeautifulSoup
browser = webdriver.Chrome()
#browser.maximize_window()

url = 'https://play.google.com/store/movies/top'
browser.get(url)

#prev_height = browser.execute_script('return document.body.scrollHeight')
while True:
    browser.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    
    curr_height = browser.execute_script('return document.body.scrollHeight')
    time.sleep(2)
    
    if curr_height == prev_height:
        break

print('스크롤 완료')

soup = BeautifulSoup(browser.page_source, 'lxml')

movies = soup.find_all('div', {'class' : 'Vpfmgd'})
print('할인 영화 모음!')
for movie in movies:
    if movie.find('span', {'class' : 'SUZt4c djCuy'}) != None:
        name = movie.find('div', {'class' : 'WsMG1c nnK0zc'}).text
        price = movie.find('span', {'class' : 'SUZt4c djCuy'}).text
        sale_price = movie.find('span', {'class' : 'VfPpfd ZdBevf i5DZme'}).text
        url = 'https://play.google.com'+movie.find('a')['href']
        print(f'영화제목 : {name}, 기존 가격 : {price}, 할인 후 가격 : {sale_price}, \n주소 {url}\n\n')
browser.close()
```
